```python
# scipy.optimize.linear_sum_assignment solve the assignment problem
from scipy.optimize import linear_sum_assignment
import numpy as np

# Read the dataset
df = pd.read_parquet('matching.parquet')

# Create an empty list to store matching results
bulk_matches = []

# Iterate over unique bulk IDs
for bulk_id in df['bulk_id'].unique():
   # Extract data for the current bulk
   bulk_df = df[df['bulk_id'] == bulk_id]

   # Assign unique indexes to orders and drivers within the bulk
   bulk_df['order_index'] = bulk_df.groupby('order_id').ngroup()
   bulk_df['driver_index'] = bulk_df.groupby('driver_id').ngroup()
   # Create an assignment matrix with the shape of (num_orders, num_drivers)
   bulk_array = np.ones((bulk_df['order_id'].nunique(), bulk_df['driver_id'].nunique()))
   for row in bulk_df[['order_index', 'driver_index', 'eta']].values:
       bulk_array[row[0], row[1]] = row[2]

   # Solve the assignment problem to find optimal matches
   # Get 2 lists with order indexes and the associated driver index respectively
   order_indexes, driver_indexes = linear_sum_assignment(bulk_array)

   # Identify assigned drivers and remove them from the main DataFrame
   assigned_drivers = bulk_df[bulk_df['driver_index'].isin(driver_indexes)]['driver_id'].tolist()
   df = df.loc[~df['driver_id'].isin(assigned_drivers)]

   # Reconstruct matching information and store it in the bulk_matches list
   for order_idx, driver_idx in zip(order_indexes, driver_indexes):
       match = bulk_df[(bulk_df['order_index'] == order_idx) & (bulk_df['driver_index'] == driver_idx)]
       order_id = match['order_id'].item()
       driver_id = match['driver_id'].item()
       eta = match['eta'].item()
       bulk_matches.append((bulk_id, order_id, driver_id, eta))
```

```python
# Create a DataFrame with the results
bulk_matches_df = pd.DataFrame(data=bulk_matches, columns=['bulk_id', 'order_id', 'driver_id', 'eta'])
print('Bulk Matching total ETA:', bulk_matches_df['eta'].sum())
```

```python
from scipy.optimize import linear_sum_assignment
import numpy as np
import pandas as pd

# Read the dataset
df = pd.read_parquet('matching.parquet')

# Create an empty list to store matching results
bulk_matches = []

# Iterate over unique bulk IDs
for bulk_id in df['bulk_id'].unique():
   # Extract data for the current bulk
   bulk_df = df[df['bulk_id'] == bulk_id].copy()  # Use .copy() to avoid the warning

   # Assign unique indexes to orders and drivers within the bulk
   bulk_df.loc[:, 'order_index'] = bulk_df.groupby('order_id').ngroup()
   bulk_df.loc[:, 'driver_index'] = bulk_df.groupby('driver_id').ngroup()

   # Create an assignment matrix with the shape of (num_orders, num_drivers)
   bulk_array = np.ones((bulk_df['order_id'].nunique(), bulk_df['driver_id'].nunique()))
   for row in bulk_df[['order_index', 'driver_index', 'eta']].values:
       bulk_array[row[0], row[1]] = row[2]

   # Solve the assignment problem to find optimal matches
   order_indexes, driver_indexes = linear_sum_assignment(bulk_array)

   # Identify assigned drivers and remove them from the main DataFrame
   assigned_drivers = bulk_df[bulk_df['driver_index'].isin(driver_indexes)]['driver_id'].tolist()
   df = df.loc[~df['driver_id'].isin(assigned_drivers)]

   # Reconstruct matching information and store it in the bulk_matches list
   for order_idx, driver_idx in zip(order_indexes, driver_indexes):
       match = bulk_df[(bulk_df['order_index'] == order_idx) & (bulk_df['driver_index'] == driver_idx)]
       order_id = match['order_id'].item()
       driver_id = match['driver_id'].item()
       eta = match['eta'].item()
       bulk_matches.append((bulk_id, order_id, driver_id, eta))
```
